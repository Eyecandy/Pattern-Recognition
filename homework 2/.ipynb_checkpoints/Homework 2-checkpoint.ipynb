2a) (50) Perform Logistic Regression like we did in class. Find $\vec{w}$

```python
features_m = []
features_p = []
classes_p = []
classes_m = []
for i in range(ndata):
    v = [1,dataxp[i],datayp[i]]
    c = [1,dataxm[i],dataym[i]]
    features_m.append(c)
    features_p.append(v)
    classes_p.append(1)
    classes_m.append(-1)

features_m = np.array(features_m)
features_p = np.array(features_p )
classes_p = np.array(classes_p)
classes_m = np.array(classes_m)

features = np.concatenate([features_p,features_m])
classes = np.concatenate([classes_p,classes_m])


def padall(x):
    tmp = []
    for i in x:
        v = [1,i]
        tmp.append(v)
    return tmp
        

class Logistic:
    def __init__(self):
        self.w = None

    def train(self, features, classes):
        def cost(w):
            s = classes*np.dot(features, w)
            s = np.clip(s,-200,200)
            t = np.log1p(1+np.exp(-s))
            return np.sum(t)
        res = minimize(cost, [0]*len(features[0]))
        self.w = res.x
        return np.array(self.w)

    def classify(self, features,w):
        s = np.dot(features, w)
        s = np.clip(s,-200,200)
        return 1./(1.+np.exp(-s))
    def classif05(self,features,w):
        s = np.dot(features, w)
        s = np.clip(s,-200,200)
    def classify2(self, features):
        s = np.dot(features, self.w)
        s = np.clip(s, -200, 200)
        return 1./(1 + np.exp(-s))
        
        
        
c = Logistic()
w = c.train(features,classes)
clsf = c.classify(features,w)

countCorrect = 0
for i in (range(ndata*2)):
    if clsf[i] > 0.5 and classes[i] == 1:
        countCorrect+=1
    elif clsf[i] < 0.5 and classes[i] ==-1:
        countCorrect+=1
countCorrect/2000.




```

2b) (50) You will find that it's quite hard to check the answer. One way to do that is to draw a line with 0.5 probability.

Draw a line of 0.5 probability along with data points. Does it match with what you expectation?

Hint: think about $\theta$ and how to get $\theta$ to return 0.5.

Hint2: it has something to do with $\vec{w}$.

Hint3: It is a straight line.

```python
def getYs(w,xs, prob):
    c, m, k = w
    
    rh = -np.log((1./prob) - 1.)
    
    ys = []
    for x in xs:
        ys.append((c + m*x + rh)/(-k)) 
    return np.array(ys)
```

```python
xs = np.linspace(-5, 5, len(features))
ys = getYs(c.w,xs, 0.5 )
```

```python
np.random.seed(1111)
ndata = 1000

#class +1
testxp = np.random.randn(ndata)+1
testyp = np.random.randn(ndata)+1

#class -1
testxm = np.random.randn(ndata)-1
testym = np.random.randn(ndata)-1

plt.figure(figsize=(8,8))
plt.plot(testxp, testyp, '.b', alpha=0.3)
plt.plot(testxm, testym, '.r', alpha=0.3)
plt.xlim(-5,4)
plt.xlim(-5,4)
plt.grid(True)
```

```python
size = 100
FN = np.zeros((1, size))
FP = np.zeros((1, size))
threshold = np.linspace(0.01, 0.99, size)
xpLen = len(testxp)
for idx in range(xpLen):
    pPos = c.classify2([1, testxp[idx], testyp[idx]])
    pNeg = c.classify2([1, testxm[idx], testym[idx]])
    for i in range(size):
        if (pNeg > threshold[i]):
            FP[0][i] += 1
        elif (pPos <= threshold[i]):
            FN[0][i] += 1
    
for i in range(size):
    FN[0][i] = FN[0][i]/len(testxp)
    FP[0][i] = FP[0][i]/len(testxp)
```
